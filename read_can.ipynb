```python
import can
import cantools
import pandas as pd
```

```python
file = can.BLFReader("logs/can/FLR1CAN_CAN6.blf")
db = cantools.database.load_file('logs/can/FLRsystem_MAIN_26_FLR1CAN_230105.dbc')
```

```python
data = []
aid = set()

for i, msg in enumerate(file):
    # if i == 1000:
    #     break
    try:
        aid.add(msg.arbitration_id)
        decoded = db.decode_message(msg.arbitration_id, msg.data)
        data.append(decoded)
    except Exception as e:
        print(str(e) + "\n\n")
```

```python
len(aid)
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('output.csv', index=False)
```
